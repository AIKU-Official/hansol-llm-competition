```python
# pip install accelerate
import accelerate
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("google/gemma-7b-it")
model = AutoModelForCausalLM.from_pretrained("google/gemma-7b-it", device_map="auto")
```

```python
import pandas as pd
from tqdm import tqdm
train_df = pd.read_csv('train.csv')
questions = train_df[['질문_1', '질문_2']]
```

```python
answers = {}
for i, row in tqdm(questions.iterrows(), total=questions.shape[0]):
    question_1 = row['질문_1']
    question_2 = row['질문_2']
    for question in [question_1, question_2]:
        #input_text = f"너는 도배, 건축물, 벽지 공사, 인테리어 영역에 대한 실무부터 이론까지 모두 파악하고 있는 전문가야. {question}에 대해 할 수 있는 한 최대한 자세하고 전문적인 답변을 작성해. '~습니다', '~입니다' 말투 대신 '~이다.' 말투로 답변을 작성하도록 해. 한국어로 작성해."
        # input_text = f"너는 도배, 건축물, 벽지 공사, 인테리어 영역에 대한 실무부터 이론까지 모두 파악하고 있는 전문가야. {question}에 대한 전문적인 답변을 작성해줘. 자세하게 부연설명을 해줘. 줄글로 아주 자세하게 써도 좋아. 토큰이나 말머리는 제거하고 문장만 작성해줘. 볼드체 또는 순번이나 키워드를 따로 작성하지 않아도 돼. 글자수를 채우지 않아도 되니 문장을 완결하여 작성해줘. "
        input_text = f"너는 도배, 건축물, 벽지 공사, 인테리어 영역에 대한 전문가야. {question}에 대해 최대한 자세하게 설명해줘. 토큰이나 말머리는 사용하지 않고, 순서나 볼드체 표시도 하지 않아도 돼. 완성된 문장으로 답변을 작성해줘."
        input_ids = tokenizer(input_text, return_tensors="pt").to("cuda")
        outputs = model.generate(**input_ids, max_new_tokens=200)
        essay = tokenizer.decode(outputs[0], skip_special_tokens=True)  # special tokens 제거
        essay = essay.replace(input_text, '').strip()
        print(essay)

        answers[question] = essay
```

```python
for question in answers:
    answers[question] = answers[question].replace('*', '').replace('\n', ' ').strip()

# DataFrame으로 변환
df_answers = pd.DataFrame(list(answers.items()), columns=['Cursed', 'Answer'])

# csv 파일로 저장
df_answers.to_csv('answers.csv', index=False, encoding='utf-8-sig')
```

```python
import json

# dict를 JSON 파일로 저장합니다.
with open('answers.json', 'w', encoding='utf-8') as f:
    json.dump(answers, f, ensure_ascii=False, indent=4)
```
